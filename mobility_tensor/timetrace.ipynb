```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the folder containing the CSV files
# folder_path = '/Users/sudhipv/documents/coupledode/codes/mobility_tensor/Stochastic_matrix'  # Change this to your folder path

folder_path = '/Users/sudhipv/documents/coupledode/codes/mobility_tensor/Stochastic_matrix_tor_ott_wind_muskoka'


delimiter = ' ' 
```

```python
# Iterate through the CSV files named from 'file0.csv' to 'file51.csv'


matrix = np.zeros((4,4,52))


for i in range(52):  # 0 to 51 inclusive
    file_name = f'Stochastic_matrix{i}.dat'
    file_path = os.path.join(folder_path, file_name)


    if os.path.isfile(file_path):
        # Read the .dat file into a DataFrame
        df = pd.read_csv(file_path, delimiter=delimiter, header=None)  # Adjust delimiter as needed

        # Check if the DataFrame is 4x4
        if df.shape == (4, 4):
            # Convert DataFrame to a numpy array
            matrix[:,:,i] = df.values
        

            # Plotting the 4x4 matrix
            # plt.figure()
            # plt.imshow(matrix, cmap='viridis', interpolation='none')
            # plt.colorbar()
            # plt.title(f'Matrix from {file_name}')
            # plt.xticks(np.arange(4))
            # plt.yticks(np.arange(4))
            # plt.show()


        else:
            print(f'{file_name} does not contain a 4x4 matrix.')
    else:
        print(f'{file_name} does not exist.')




print(matrix[:,:,3])
```
